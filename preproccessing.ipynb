```python
# 1. Mengimpor Library yang Diperlukan
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from google.colab import files
# 2. Mengunggah File Dataset
print("Silakan unggah file 'drug_consumption.data'")
uploaded = files.upload()
nama_file_data = next(iter(uploaded))
print(f"File '{nama_file_data}' berhasil diunggah.")
print("-" * 40)
# 3. Memuat Dataset
nama_kolom = [
    'ID', 'Usia', 'JenisKelamin', 'Pendidikan', 'Negara', 'Etnis', 'Nscore',
    'Escore', 'Oscore', 'Ascore', 'Cscore', 'Impulsif', 'SS', 'Alkohol',
    'Amfetamin', 'Amil', 'Asph', 'Benzos', 'Kafein', 'Ganja', 'Cokelat', 'Kokain',
    'Crack', 'Ekstasi', 'Heroin', 'Ketamin', 'Legalh', 'LSD', 'Meth',
    'Jamur', 'Nikotin', 'Semer', 'VSA'
]
dataset = pd.read_csv(nama_file_data, header=None, names=nama_kolom)
X = dataset.iloc[:, 1:13].values
y = dataset.iloc[:, 19].values
print("--- Data Awal Berhasil Dimuat ---")
print("Contoh 5 baris pertama Fitur (X):")
print(X[:5])
print("\nContoh 5 baris pertama Target (y) sebelum di-encode:", y[:5])
print("-" * 40)

# 4. Mengkodekan Data Kategori (Variabel Target)
le = LabelEncoder()
y = le.fit_transform(y)

print("--- Setelah Encoding ---")
print("Contoh 5 baris pertama Target (y) setelah di-encode:", y[:5])
print("-" * 40)

# 5. Membagi Dataset menjadi Data Latih dan Data Uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print("--- Setelah Pembagian Dataset ---")
print("Jumlah data latih (X_train):", X_train.shape[0])
print("Jumlah data uji (X_test):", X_test.shape[0])
print("-" * 40)

# 6. Penskalaan Fitur (Feature Scaling)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print("--- Setelah Penskalaan Fitur ---")
print("Contoh 5 baris pertama X_train setelah penskalaan:")
print(X_train[:5])
print("\nPra-pemrosesan di Google Colab Selesai! ✅")
```
